Limpieza de texto

```python
import re

def limpiarTexto(text):
    text = re.sub(r'[-,;+*?¿!¡<>{}// %&´""“”–°º−()]',' ',text)
    text = re.sub('^ +',' ',text)
    text = re.sub('\n ','',text)
    text = re.sub('\n\n+','\n',text)
    text = re.sub(' +\n','\n',text)
    text = re.sub('\. \n','\.\n',text)
    text = re.sub('\d\n','',text)
    text = re.sub('\x0c','',text)
    text = re.sub('¨[... ]','',text)
    text = re.sub(' {2,100}',' ',text)
    text = re.sub('^ +','',text)
    text = re.sub('n.º.','',text)
    text = re.sub('[a-zA-z-.]@[a-zA-Z].com','',text)
    text = re.sub('\d+','',text)

    text = re.sub(' +\n+','\n',text)
    text = re.sub('\n+','\n',text)
    text = re.sub('\d+\n','',text)
    text.strip()
    return text

df1 = df.apply(str)
df1 = df.apply(limpiarTexto)
df1 = df1.head(20)
  
```

```python
#Normalización
def normalizar(text):
    # todo a minúsculas
    text = text.lower()

    # tildes y diacríticas
    text = re.sub('á', 'a', text)
    text = re.sub('é', 'e', text)
    text = re.sub('í', 'i', text)
    text = re.sub('ó', 'o', text)
    text = re.sub('ú', 'u', text)
    text = re.sub('ü', 'u', text)
    text = re.sub('ñ', 'n', text)

    return text

dfNorm = df1.apply(normalizar)
dfNorm[1]
```

```python
import nltk

stops = nltk.corpus.stopwords.words('spanish')

#elimino stopwords
def remove_stops(texto: str) -> str:
    """
    Función que elimina stopwords
    Params:
        **texto**:texto a ser limpiado de stopwords

    """
    texto = [
        i for i in texto.split() if i not in stops
    ]
    return texto

df1.apply(str)
df1 = dfNorm.apply(remove_stops)
```
