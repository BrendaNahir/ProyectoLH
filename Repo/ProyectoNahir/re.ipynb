#PROBANDO CON UNA PERICIA

```python
import pandas as pd
import numpy as np

df = pd.read_csv('pericias_medicas.csv', sep=',',  encoding='utf-8')
#convierte en string los datos de la columna text
df = df['text'].apply(str)

print(df)
```

```python
#Se guarda el texto de la primer pericia
df1= df[0]
df1
```

```python
#Elimina el espacio al principio y final del texto
df2 = df1.strip()
(df2)
```

```python
#Hace un salto de linea después de un "." y "\n"
import re

df3 = re.split(r'\. \n', df2)
df3
```

```python
#Elimina los espacios duplicados
import re

df4 =" ".join(re.split("\s+", df3[1], flags=re.UNICODE))
df4
#Al hacer esto me muestra por párrafo (df2[0] primer párrafo, df2[1] segundo párrafo y así.)
```

```python
import re

df5 = (re.findall(r"[A-Z]\w+\:| :", df4))  
df5
```

TFIDF (VECT)

```python
#Se importan las librerias necesarias y se toman los datos de muestra
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer 
df4 
#Será que tengo que hacer con varios párrafos?? o con el dataset entero????
```

```python
#Convierte los datos de muestra en DataFrame usando Pandas
df6 = pd.DataFrame({'Un párrafo': [df4]})
print (df6)
```

```python
#Inicializa el vectorizador
#Se inicializa el vectorizador y se ajusta y transforman los datos
tfidf_vectorizer = TfidfVectorizer()
doc_vec = tfidf_vectorizer.fit_transform(df6.iloc[0])
```

```python
#Convierte los datos transformados en un marco de datos
df7 = pd.DataFrame(doc_vec.toarray().transpose(), index= tfidf_vectorizer.get_feature_names_out())
```

```python
#Cambia los nombres de las columnas e imprime el resultado
df7.columns = df6.columns
print(df7)
```
